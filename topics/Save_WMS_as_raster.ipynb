```python
import leafmap
from samgeo import tms_to_geotiff
```

```python
m = leafmap.Map(center=[40.78, -73.95], zoom=13)
```

```python
source_url = "https://mt1.google.com/vt/lyrs=y&x={x}&y={y}&z={z}"
```

```python
m.add_tile_layer(
    url=source_url,
    name = "Google Satellite",
    attribution="Google"
)
```

```python
m
```

```python
bbox = m.user_roi_bounds()
```

```python
bbox
```

```python
image = "crop.tif"
```

```python
tms_to_geotiff(
    output=image,
    bbox=bbox,
    zoom=20,
    source=source_url,
    overwrite=True
)
```

```python
from rasterio.plot import show
import rasterio
```

```python
with rasterio.open(image) as src:
  show(src)
```
